## Should we use time series model? or regression model?

To determine whether we need a time series prediction model like ARIMA or Prophet, or whether a simpler machine learning regression model like SVM or XGBoost regression is appropriate, you can follow a systematic process of exploratory data analysis and preliminary modeling:

### 1. **Understand the Nature of The Data**
   - **Stationarity**: Check if the target variable \( C \) is stationary. If it shows trends or seasonality, time series models may be needed. In our case our target would be the GDP and Interest Rate column
   - **Temporal Dependencies**: Analyze autocorrelations and partial autocorrelations for \( C \). If significant lags are identified, it suggests that past values influence future values, favoring time series models.

### 2. **Examine Relationships and Features**
   - **Cross-Correlation**: Determine if thereâ€™s a time-lagged relationship between features (\( A \) and \( B \)) and the target (\( C \)). Strong cross-correlations at specific lags suggest a time series approach.
   - **Feature Behavior**: If features \( A \) and \( B \) themselves show patterns or trends over time, models that can incorporate temporal dynamics might be more suitable.

### 3. **Initial Model Fitting**
   - **Baseline Models**: Fit simple linear models or basic machine learning models (like linear regression) to see how well they perform without considering time series aspects.
   - **Time Series Models**: Fit basic time series models (like ARIMA) to the univariate series \( C \) to see if these models capture the data dynamics better.

### 4. **Performance Comparison**
   - **Residual Analysis**: After fitting initial models, analyze residuals for any patterns. If residuals from simple models show structure or autocorrelation, it suggests that these models are not capturing all the temporal information.
   - **Validation Metrics**: Use time-based cross-validation to compare model performances. Metrics such as RMSE, MAE, and AIC can be instrumental in deciding which model handles the data dynamics better.

### 5. **Complexity vs. Performance Trade-off**
   - **Model Complexity**: Consider the complexity of implementing and maintaining models like ARIMA or Prophet versus simpler ML models like SVM or XGBoost.
   - **Scalability and Updating**: Time series models might require regular re-fitting as new data becomes available, whereas machine learning models might be more robust to changes over time without frequent re-training.

### 6. **Domain Considerations**
   - **Interpretability**: Depending on the need for understanding how input features affect the output, simpler models or models like XGBoost that provide feature importance might be preferable.
   - **Data Availability**: Time series models generally need a longer historical data series to be effective, while machine learning models can sometimes work with shorter spans.

### Decision Making
- **Time Series Model Needed**: If your data analysis and initial modeling suggest significant time dependencies and seasonal patterns, and these aspects are crucial for your predictions, opt for a time series model.
- **Machine Learning Regression Suitable**: If the target variable does not show strong time dependencies, or if the relationships are adequately captured by the features at their current values, a machine learning regression model might be more appropriate.

Parameter Tuning

```python
from scipy.stats import uniform
from prophet import Prophet
from mango import scheduler, Tuner
from sklearn.metrics import mean_squared_error
import numpy as np
import traceback

global MANGO_train_data
global MANGO_feature_list 
global MANGO_prediction_date
global MANGO_future_regressors
global MANGO_local_final_data
global MANGO_yname

param_space = dict(growth = ['linear', 'flat'],
                #    n_changepoints  = range(0, 55, 5),
                #    changepoint_range  = uniform(0.5, 0.5),
                   yearly_seasonality = [True, False],
                   weekly_seasonality = [True, False],
                   daily_seasonality = [True, False],
                   seasonality_mode = ['additive', 'multiplicative'],
                   seasonality_prior_scale=uniform(5.0, 15.0),
                   changepoint_prior_scale=uniform(0.0, 0.1),
                   interval_width = uniform(0.2, 0.8),
                   uncertainty_samples = [500, 1000, 1500, 2000]
                   )


def objective_function(args_list): 
    params_evaluated = []
    results = []
    
    for params in args_list:
        try:
            model = Prophet(**params)
            
            for feature in MANGO_feature_list:
                model.add_regressor(feature)

            model.fit(MANGO_train_data)
            
            future = model.make_future_dataframe(periods=1, include_history=False, freq='MS')
            # Merge future regressors and training data
            future = future.merge(MANGO_future_regressors,on='ds',how='left')
            future = future.merge(MANGO_train_data,on='ds',how='outer')
            future = future.rename(columns={'Date': 'ds', MANGO_yname: 'y', 'Interest_Rate_x' : 'Interest_Rate'})
            future = future[future['ds'] == MANGO_prediction_date]
            actual = MANGO_local_final_data[MANGO_local_final_data['ds'] == MANGO_prediction_date]

            forecast = model.predict(future)

            rmse = np.sqrt(mean_squared_error(actual[actual['ds'] == MANGO_prediction_date]['y'],forecast[forecast['ds'] == MANGO_prediction_date]['yhat']))
            
            params_evaluated.append(params)
            results.append(rmse)
        except Exception as e:
            print(f"Exception raised for {e}")
            print(traceback.format_exc())
            pass 
            params_evaluated.append(params)
            results.append(25000.0)# Giving high loss for exceptions regions of spaces
        
    return params_evaluated, results
```

```python
import pandas as pd
import numpy as np
import json
from prophet import Prophet
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

import logging
logging.getLogger('cmdstanpy').setLevel(logging.WARNING) 

def plot_historical(actual_total, forecast_total, rmse_values, average_rmse, yname, output_dir, target_pretty_name, filename, annotate_text=True):
    plt.figure(figsize=(10, 6))
    plt.plot(actual_total['ds'], actual_total['y'], label='Actual')
    plt.plot(forecast_total['ds'], forecast_total['yhat'], label='Forecast', marker='o')

    # Draw lines between actual and predicted points and annotate RMSE
    for i, date in enumerate(forecast_total['ds']):
        # Ensure there's an actual data point to match the forecast date
        if date in actual_total['ds'].values:
            actual_y = actual_total[actual_total['ds'] == date]['y'].values[0]
            forecast_y = forecast_total[forecast_total['ds'] == date]['yhat'].values[0]
            plt.plot([date, date], [actual_y, forecast_y], 'k--')  # 'k--' denotes a black dashed line

            # Calculate mid-point for RMSE annotation
            mid_y = (actual_y + forecast_y) / 2
            if annotate_text:
                plt.annotate(f'RMSE: {rmse_values[i]:.2f}', (date, mid_y), textcoords="offset points", xytext=(0,10), ha='center')

    # Annotate average RMSE
    plt.annotate(f'Average RMSE: {average_rmse:.2f}', (0.05, 0.95), xycoords='axes fraction', fontsize=12, bbox=dict(boxstyle="round,pad=0.3", fc="yellow", ec="black", lw=1))

    plt.title(f'{target_pretty_name} Forecast vs Actual with RMSE Values', fontsize=10)
    plt.xlabel('Date')
    plt.ylabel(f'{target_pretty_name}')
    plt.legend()
    plt.savefig(f'{output_dir}/{filename}')
    plt.close()

def forecast_and_plot_quarterly(final_data, feature_list, prediction_dates, yname, target_pretty_name, output_dir):
    """
    Trains and predicts GDP at specified dates using available historical data up to the last prediction,
    and saves the plots for each prediction.

    Parameters:
    - final_data (DataFrame): The input data including dates,GDP,and potential regressors.
    - feature_list (list): A list of column names to be used as regressors.
    - prediction_dates (list): A list of dates (str) when predictions are required.
    - output_dir (str): Directory to save the output plots.
    """

    actual_total = pd.DataFrame()
    forecast_total = pd.DataFrame()

    rmse_values = []

    # Ensure data is in the correct format
    local_final_data = final_data.rename(columns={'Date': 'ds', yname: 'y'})

    forecast = None

    for date in prediction_dates:
        prediction_date = pd.to_datetime(date)
        train_end = prediction_date - pd.DateOffset(months=1)
        regressors_end = prediction_date

        print(f'Training model for {prediction_date}...')

        # Prepare data for this cycle
        train = local_final_data[(local_final_data['ds'] <= train_end)]
        future_regressors = local_final_data[(local_final_data['ds'] > train_end) & (local_final_data['ds'] <= regressors_end)][['ds'] + feature_list]

        # Modelling Part! With Parameter Tuning 
        conf_Dict = dict()
        conf_Dict['initial_random'] = 2
        conf_Dict['num_iteration'] = 2

        global MANGO_train_data
        global MANGO_feature_list
        global MANGO_prediction_date
        global MANGO_future_regressors
        global MANGO_yname
        global MANGO_local_final_data

        MANGO_train_data = train.copy()
        MANGO_feature_list = feature_list.copy()
        MANGO_prediction_date = prediction_date
        MANGO_future_regressors = future_regressors.copy()
        MANGO_local_final_data = local_final_data.copy()
        MANGO_yname = yname
        

        tuner = Tuner(param_space, objective_function, conf_Dict)
        results = tuner.minimize()
        print('best parameters:', results['best_params'])
        print('best loss:', results['best_objective'])

        # Saving the best parameters to a JSON file
        with open(f'PT_experiment_2/params/best_parameters_{yname}_{prediction_date.strftime("%y-%m")}.json', 'w') as json_file:
            json.dump(results['best_params'], json_file, indent=4)

        # Train based on best parameters
        model = Prophet(**results['best_params'])
        
        for feature in feature_list:
            print(feature)
            model.add_regressor(feature)

        model.fit(train)

        # Create future dataframe and predict
        future = model.make_future_dataframe(periods=1, include_history=False, freq='MS')

        # Merge future regressors and training data
        future = future.merge(future_regressors,on='ds',how='left')
        future = future.merge(train,on='ds',how='outer')

        # Combine columns from both merges where '_x' and '_y' columns may exist
        for feature in feature_list:
            future[feature] = future[f'{feature}_x'].combine_first(future[f'{feature}_y'])
            future.drop([f'{feature}_x',f'{feature}_y'],axis=1,inplace=True)

        forecast = model.predict(future)

        # Plotting and comparing the forecast to actual data
        actual = local_final_data[(local_final_data['ds'] >= train_end) & (local_final_data['ds'] <= regressors_end)]
        forecast = forecast[forecast['ds'] == regressors_end]

        actual_total = pd.concat([actual_total,actual])
        forecast_total = pd.concat([forecast_total,forecast])
        
        rmse = np.sqrt(mean_squared_error(actual[actual['ds'] == regressors_end]['y'],forecast[forecast['ds'] == regressors_end]['yhat']))
        rmse_values.append(rmse)

    fig = model.plot_components(forecast)
    fig.savefig(f'PT_experiment_2/{yname}_forecast_components.png')

    # Calculate average RMSE
    average_rmse = np.mean(rmse_values)

    # Plotting Without Historical data
    plot_historical(actual_total,forecast_total,rmse_values,average_rmse,yname,output_dir,target_pretty_name,f"{yname}_retail_forecast_rmse.png")

    # Plotting number with historical data
    local_final_data = local_final_data[local_final_data['ds'] < regressors_end]
    actual_total = pd.concat([actual_total,local_final_data])
    
    actual_total = actual_total.sort_values('ds')
    actual_total = actual_total.drop_duplicates()
    
    plot_historical(actual_total,forecast_total,rmse_values,average_rmse,yname,output_dir,target_pretty_name,f"{yname}_hist_retail_forecast_rmse.png",annotate_text=False)

    return model

features = [
    "Interest_Rate"
]

to_simulate = {
    'Retail Sales Survey: Retail Sales Index: Food, Drinks, and Tobacco': 'RS_FoodDrinkTobacco',
    'Retail Sales Survey: Retail Sales Index: Other Goods: Apparel': 'RS_Apparel',
    'Retail Sales Survey: Retail Sales Index: Household Appliances': 'RS_Appliances',
    'Retail Sales Survey: Retail Sales Index: Stationery & Communication': 'RS_StationeryComm',
    'Retail Sales Survey: Retail Sales Index: Cultural & Recreation': 'RS_CultureRec',
    'Retail Sales Survey: Retail Sales Index: Fuels': 'RS_Fuels',
    'Retail Sales Survey: Retail Sales Index: CarSpareParts&Accessories': 'RS_CarParts',
    'Retail Sales Survey: Retail Sales Index: Other Goods': 'RS_OtherGoods',
    'Hotel Room Occupancy: Jakarta: Average': 'Hotel_Jakarta_Avg',
    'Hotel Room Occupancy: Bali: Average': 'Hotel_Bali_Avg',
    'Electronic Card Transaction: Credit Card: Value': 'Card_Credit_Val',
    'Electronic Card Transaction: ATM and Debit Card: Value': 'Card_Debit_Val',
    'Consumption: Cement: Domestic': 'Cons_Cement',
    'Construction Machinery: Sales Volume: United Tractors (Komatsu): Total': 'ConstMach_Sales',
    'Motor Vehicle Sales: Wholesales': 'MV_Wholesale',
}

# Example of how to call the function
prediction_dates = [
    "1/1/2023",
    "2/1/2023",
    "3/1/2023",
    "4/1/2023",
    "5/1/2023",
    "6/1/2023",
    "7/1/2023",
    "8/1/2023",
    "9/1/2023",
    "10/1/2023",
    "11/1/2023",
    "12/1/2023",
    "1/1/2024",
    "2/1/2024",
    "3/1/2024",
    "4/1/2024",
]

for target_pretty_name, target in to_simulate.items():
    model = forecast_and_plot_quarterly(df_all, features, prediction_dates, target, target_pretty_name, 'PT_experiment_2')
```
